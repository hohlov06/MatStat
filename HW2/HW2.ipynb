# 2.1

Имеются: выборка $(X_1, ..., X_{100})$, распределение Парето $\alpha > 0 $

$p(x)=\dfrac{\alpha}{x^{\alpha+1}}$, $x > 1$

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return m-h, m, m+h
```

```python
alphaArray = np.linspace(0.01, 4.99, 100)
leftArray = []
rightArray = []
centralArray = []

for alpha in alphaArray:
    r = stats.pareto.rvs(alpha, size=100)
    left,central, right = mean_confidence_interval(r)
    leftArray.append(left)
    rightArray.append(right)
    centralArray.append(central)
```

```python
plt.figure(figsize=(10, 6))

plt.semilogy(alphaArray, leftArray,label="left stat")
plt.xlabel("alpha")
plt.ylabel("stat")
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

plt.semilogy(alphaArray, rightArray, label= "right stat")
plt.xlabel("alpha")
plt.ylabel("stat")
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

plt.semilogy(alphaArray, np.array(rightArray) - np.array(leftArray), label= "interv")
plt.xlabel("alpha")
plt.ylabel("stat")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(alphaArray, leftArray,label="left stat")
plt.plot(alphaArray, rightArray, label= "right stat")
plt.plot(alphaArray,centralArray, label= "stat")
plt.fill_between(alphaArray, leftArray, rightArray, alpha=0.5)
plt.xlabel("alpha")
plt.ylabel("stat")
plt.ylim(-1, 20)
plt.legend()
plt.show()
```

Выводы: интеграл матожидания для распеределния Парето расходится

# 2.3 
$N(\theta,1)$

$p(x) = \dfrac{1}{\sqrt{2\pi}} e^{-\frac{(x-\theta)^2}{2}}$

```python
Nsize = 100
theta_0 = Number_in_list

k_array = np.linspace(1, 5, 21)
```

```python
for theta in [theta_0 * k for k in k_array]:
    sample = stats.norm(theta, 1).rvs(size=Nsize)
    c_alpha = stats.norm(Nsize * theta_0, Nsize).ppf(1 - 0.05)
    print('theta = %d: sample_sum = %.2f, c_alpha = %.2f' % (theta, sample.sum(), c_alpha))
```
